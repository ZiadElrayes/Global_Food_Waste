1Ô∏è‚É£ Country

üìå Definition: The name of the country where the food wastage data was recorded.

üîç Insight: Helps identify which countries have the highest or lowest food waste levels.

2Ô∏è‚É£ Year

üìå Definition: The year in which the food wastage data was collected (ranging from 2018 to 2024).

üîç Insight: Allows for trend analysis to see how food waste has evolved over time.

3Ô∏è‚É£ Food Category

üìå Definition: The type of food that was wasted (e.g., fruits, vegetables, dairy, meat, grains).

üîç Insight: Helps determine which food categories contribute the most to overall wastage.

4Ô∏è‚É£ Total Waste (Tons)

üìå Definition: The total amount of food wasted in a country for a given year, measured in metric tons.

üîç Insight: Provides an understanding of the scale of food wastage in absolute terms.

5Ô∏è‚É£ Economic Loss (Million $)

üìå Definition: The estimated financial loss due to food wastage, measured in millions of U.S. dollars.

üîç Insight: Shows the financial impact of food waste on a country‚Äôs economy.

6Ô∏è‚É£ Avg Waste per Capita (Kg)

üìå Definition: The average amount of food wasted per person in a country, measured in kilograms.

üîç Insight: Highlights how much food is wasted on an individual level, useful for comparisons across countries.

7Ô∏è‚É£ Population (Million)

üìå Definition: The total population of the country in the given year, measured in millions.

üîç Insight: Helps normalize waste data and analyze per capita trends.

8Ô∏è‚É£ Household Waste (%)

üìå Definition: The percentage of total food waste that comes from households (as opposed to retail, restaurants, or farms).

#### 2 ) Which countries have the highest and lowest food wastage levels?

```python
country_tw = gfw.groupby('Country')['Total Waste (Tons)'].sum().sort_values(ascending = False)
country_tw
```

```python
# Group by Year and Country, then sum the Total Waste (Tons)
yearly_waste = gfw.groupby(["Year", "Country"])["Total Waste (Tons)"].sum().reset_index()

# Find the country with the maximum waste for each year
most_wasting_countries = yearly_waste.loc[yearly_waste.groupby("Year")["Total Waste (Tons)"].idxmax()]

# Display the results
print(most_wasting_countries)
```

```python
colors = sns.color_palette("husl", len(country_tw))
plt.figure(figsize = (25, 6))
plt.bar(country_tw.index , country_tw.values , color = colors)
plt.xlabel("Country")
plt.ylabel("Total Waste of food in Tons")
plt.title("Total waste of Food in Tons for each Country through the years from 2018 - 2024")
```

#### 9 ) Based on historical data, which regions are projected to have the highest food waste in the coming years?

```python
country_region = {
    "USA": 'North America',
    "China": 'Asia',
    "Japan": 'Asia',
    "Germany": 'Europe',
    "India": 'Asia',
    "UK": 'Europe',
    "France": 'Europe',
    "Italy": 'Europe',
    "South Africa": 'AFC',
    "Canada": 'North America',
    "South Korea": 'Asia',
    "Russia": 'Asia',
    "Argentina":"South America",
    "Brazil": 'South America',
    "Australia": 'Ocean',
    "Spain": 'Europe',
    "Mexico": 'North America',
    "Indonesia": 'Asia',
    "Netherlands": 'Europe',
    "Saudi Arabia": 'Asia',
    "Turkey": 'Asia',
    "Switzerland": 'Europe' 
}
```

```python
gfw['Country_Region'] = gfw['Country'].map(country_region)
gfw.head()
```

```python
selected_years = [2018, 2020, 2024]
gfw_filtered = gfw[gfw["Year"].isin(selected_years)]
```

```python
# Reset index to convert into a DataFrame
cr = gfw_filtered.groupby('Country_Region')['Country'].value_counts().reset_index(name='Count')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=cr, x='Country_Region', y='Count', hue='Country', dodge=False)

# Formatting
plt.title("Number of Countries per Region")
plt.xlabel("Region")
plt.ylabel("Number of Countries")
plt.legend(title="Countries", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Group by 'Country_Region' and sum up 'Total Waste (Tons)'
region_waste = gfw_filtered.groupby('Country_Region')['Total Waste (Tons)'].sum().reset_index()

# Sort for better visualization
region_waste = region_waste.sort_values(by='Total Waste (Tons)', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=region_waste, x='Country_Region', y='Total Waste (Tons)', palette="viridis")

# Formatting
plt.xticks(rotation=45, ha='right')
plt.title("Total Food Waste (Tons) by Region")
plt.xlabel("Region")
plt.ylabel("Total Waste (Tons)")

plt.show()
```
